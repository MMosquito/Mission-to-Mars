````python
# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
````

````python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
````

````python
# Visit the Mars NASA news site: https://redplanetscience.com
url = "https://redplanetscience.com"
browser.visit(url)
````

### Step 2: Scrape the Website

Create a Beautiful Soup object and use it to extract text elements from the website.

````python
# Create a Beautiful Soup object
html = browser.html
title_preview_soup = soup(html, "html.parser")
````

````python
# Extract title & previews together (first article)
text = title_preview_soup.select_one("div.list_text")
text
````

````python
# Extract title & previews together (all articles)
title_previews = title_preview_soup.find_all("div", class_="list_text")
for title_preview in title_previews:
    print(title_previews)
````

Extract the titles and preview text of the news articles that you scraped. Store the scraping results in Python data structures as follows:

````python
# Extract the title and preview text from the elements
title = text.find("div", class_="content_title").text
article = text.find("div", class_="article_teaser_body").text
print('---------------')
print(title)
print(article)
````

````python
# Print multiple title and previews on page
title_previews = title_preview_soup.find("div",{"id": "news"})
for title_preview in title_previews:
    print(title_previews.text)
````

````python
# Create a dictionary for title & previews
dict_title_preview = {"preview: " + article, "title: " + title}
print(dict_title_preview)
````

````python
# Add the dictionary to the list
text = title_preview_soup.find_all('div', class_='list_text')

results = []

for row in text:
    title = row.find('div', class_='content_title').get_text()
    preview = row.find('div', class_='article_teaser_body').get_text()
    result = {'title':title, 'preview':preview}
    results.append(result)
````

````python
# Print the list to confirm success
results
````
